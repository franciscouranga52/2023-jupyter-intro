## Additional steps
### Re-import the database
### Remove commas from the 'Median household income column'
### Convert to numeric

```python
#Re-import the database
df = pd.read_csv("data/cable-bill-only.csv")
df
```

```python
# Remove commas from the 'Median household income column'
df['Median household income'] = pd.to_numeric(df['Median household income'].str.replace(',', ''))
df
```

```python
### Calculate percentage change with respect to 1995: (new data - 1995 data) / 1995 data * 100 
df["cable_bill_pct_change"] = (df["Avg. cable bill"] / df["Avg. cable bill"].iat[0] -1) *100
df["household_income_pct_change"] = (df["Median household income"] / df["Median household income"].iat[0] -1) *100
df["cost-of-living_pct_change"] = df["Social Security cost-of-living adjustment"] - df["Social Security cost-of-living adjustment"].iat[0] 
df
```

```python
# Create a graphic
df.plot(x="Year",
       y=["cable_bill_pct_change", "household_income_pct_change", "cost-of-living_pct_change"],
       title = "Cable bill v. income and cost of living (Percentage change compared to 1995)",
       style = {"cable_bill_pct_change":"red",
               "household_income_pct_change":"blue",
               "cost-of-living_pct_change":"gray"},
       legend = {"cable_bill_pct_change":"Cable bill",
               "household_income_pct_change":"Household income",
               "cost-of-living_pct_change":"Cost of living"}
       )
```
